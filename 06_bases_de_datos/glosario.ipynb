### ¿Que son las entidades?

* La *entidad* es un <u>objeto unico</u> en el mundo real
 - Clientes de una empresa
 - Autos que son vendidos
 - Departamento de ventas
 - Vendedores
 - Abstractas o tangibles. se representa por un *rectangulo*

### Tipos de Entidad

Ejemplo:
* Empleado y Dependiente del Empleado

- una entidad fuerte debe contener un atributo exclusivo que es la **Clave Principal**
- una entidad debil tendria una <u>clave parcial</u> que se puede complementar con la clave principal de la entidad fuerte.

### Definiendo entidades(fuertes y debiles) en el DER

- resaltando en negrita las entidades
- resaltando en amarillo las entidades fuetes y en azul las entidades debiles

### Tipos de Relacion:

* las relaciones se representan a través del rombo
* grados de relacion: 
    - Binaria: trabaja en dos entidades
    - Ternaria: trabaja en tres entidades
    - N-aria: multiples entidades
* Concepto de Cardinalidad(Binaria, Ternaria, N-aria)

### Tipos de Cardinalidad

* tipos: (1,1), (0,1), (1,N), (0,N), (M,N)

### Tipos de Atributos

* Atributo Simple
    es un atributo indivisible, un atributo atómico, ej: dni.
* Atributo Compuesto
    puede ser dividido en partes menores que representan a otros atributos, ej: direccion(ciudad, estado, calle, ...)
* Atributo Multivalor
    puede tener uno a N(varios) valores asiciados a el, ej: telefono
* Atributo Derivado y Almacenado
    los derivados dependen de otro atributo o incluso de otra entidad para existir, ej: edad y fecha de nacimiento.
    los almacenados son los derivados de otro atributo como edad que depende de la fecha de nacimiento
* Atributo Clave
    identificar de forma unica una entidad, ej: codigo de producto
