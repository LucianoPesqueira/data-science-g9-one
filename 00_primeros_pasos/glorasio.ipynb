```python
#variables numericas
cantidad_vigilantes = 5
salario_vigilantes = 300

cantidad_docentes = 16
salario_docentes = 500

cantidad_coordinadores = 2
salario_coordinadores = 600

#1-cantidad total de empleados
print(f"La cantidad total de empleados es: {cantidad_vigilantes + cantidad_docentes + cantidad_coordinadores}")

#2-diferencia entre el salario mas bajo y mas alto
print(f"La diferencia de salario es: ${salario_coordinadores - salario_vigilantes}")

#3-promedio ponderado de los salarios de los empleados de la escuela
cantidadEmpleados = cantidad_vigilantes + cantidad_docentes + cantidad_coordinadores
total_vigilantes = cantidad_vigilantes * salario_vigilantes
total_docentes = cantidad_docentes * salario_docentes
total_coordinadores = cantidad_coordinadores * salario_coordinadores
print(f"El promedio es: ${round((total_vigilantes + total_docentes + total_coordinadores) / cantidadEmpleados, 2)}")
```

#### WHILE

```python
#ciclo WHILE
contador = 1
while contador <= 10:
    print(contador)
    contador += 1
```

```python
contador = 1
while contador <= 3:
    nota_1 = float(input("Ingrese la primera nota: "))
    nota_2 = float(input("Ingrese la segunda nota: "))
    print(f"El promedio del alumno {contador} es: {nota_1 + nota_2 / 2}")
    contador += 1
```

```python
for n in range(1, 4):
    nota_1 = float(input("Ingrese la primera nota: "))
    nota_2 = float(input("Ingrese la segunda nota: "))
    print(f"El promedio del alumno {n} es: {nota_1 + nota_2 / 2}")
```

## Estructuras de Datos

### LISTAS

```python
#coleccion: LISTAS
lista = ["Luciano Pesqueira", 9.5, 9.0, 9.7, True]
lista[0]#str
lista[1]#float
lista[-1]#bool

for elemento in lista:
    print(elemento)

lista[2] = 10.0

promedio = (lista[1] + lista[2] + lista[3]) / 3
print(promedio)
```

```python
#manipulacion de listas
lista = ["Luciano Pesqueira", 9.5, 9.0, 9.7, True]
print(len(lista))
print(lista[0:2])
print(lista[-1])
lista.append(promedio)
print(lista)
lista.extend([10.0, 8.0, 9.0])
print(lista)
lista.remove(9.0)#remueve la primera ocurrencia
print(lista)
```

### DICCIONARIO

```python
diccionario = {
    'matricula' : 2000168933,
    'dia_registro' : 25,
    'mes_registro' : 10,
    'grupo' : '2E'}
print(diccionario)
print(diccionario['matricula'])
diccionario['grupo'] = '2G'
print(diccionario)
diccionario['modalidad'] = 'EAD'
print(diccionario)
```

```python
#Profundizando en diccionarios
#pop()
diccionario.pop('grupo') if 'grupo' in diccionario else 'pass'
print(diccionario)
#items()
print(diccionario.items())
#keys()
print(diccionario.keys())
#values()
print(diccionario.values())
```

```python
#diccionario con for
for llave, value in diccionario.items():
    print(llave, value)
```
