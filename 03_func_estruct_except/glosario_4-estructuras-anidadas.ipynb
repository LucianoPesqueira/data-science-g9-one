````python
def promedio(lista: list=[0]) ->float:
    '''Funcion para calcular el promedio de notas de una lista  

    lista: list, default [0]  
      Lista con las notas para calcular el promedio  
    return = calculo: float  
      Promedio Calculado  
    '''
    calculo = sum(lista) / len(lista)
    
    return calculo
````

````python
notas = [[8.0, 9.0, 10.0], [9.0, 7.0, 6.0], [3.4, 7.0, 7.0], [5.5, 6.6, 8.0], [6.0, 10.0, 9.5]]

promedios = [round(promedio(nota),1) for nota in notas]
print(promedios) #[9.0, 7.3, 5.8, 6.7, 8.5]
````

## List comprehension con if y else
- ```[resultado_if if condicion else resultado_else for item in lista]```

````python
nombres = [('Juan', 'J584'), ('Maria', 'M788'), ('Jose', 'J975'), ('Claudia', 'C674'), ('Ana', 'A247')]
notas = [[8.0, 9.0, 10.0], [9.0, 7.0, 6.0], [3.4, 7.0, 7.0], [5.5, 6.6, 8.0], [6.0, 10.0, 9.5]]
promedios = [9.0, 7.3, 5.8, 6.7, 8.5]

situacion = ["Aprobado" if promedio >= 7 else "Reprobado" for promedio in promedios]

print(situacion)
````

````python
registros = [x for x in [nombres, notas, promedios, situacion]]
lista_completa = [nombres, notas, promedio, situacion]
print(lista_completa)
````

## Dict comprehension
- ```{llave: valor for item in lista}```

````python
lista_completa = [
    [('Juan', 'J584'), ('Maria', 'M788'), ('Jose', 'J975'), ('Claudia', 'C674'), ('Ana', 'A247')], 
    [[8.0, 9.0, 10.0], [9.0, 7.0, 6.0], [3.4, 7.0, 7.0], [5.5, 6.6, 8.0], [6.0, 10.0, 9.5]],
    [9.0, 7.3, 5.8, 6.7, 8.5],
    ['Aprobado', 'Aprobado', 'Reprobado', 'Reprobado', 'Aprobado']]

columnas = ['Notas', 'Promedio Final', 'Situacion']

registro = {columnas[i]:lista_completa[i+1] for i in range(len(columnas))}

print(registro)

registro['Estudiante'] = [lista_completa[0][i][0] for i in range(len(lista_completa[0]))]
print(registro)
print(lista_completa[0])
````

````python
#10
#diccionario -> clave(nombre_estado_unicos): valores(num empleados) otro dict clave(nombre_estado_unico): valores(suma_estados_unicos)
empleados = [('CMX', 16), ('OAX', 8), ('PUE', 9), ('PUE', 6), ('CMX', 10), 
             ('PUE', 4), ('OAX',9),  ('OAX', 7), ('OAX', 12), ('CMX', 7), 
             ('CMX', 11), ('PUE',8), ('OAX',8), ('CMX',9), ('VER', 13), 
             ('PUE', 5),  ('VER', 9), ('CMX', 12), ('PUE', 10), ('CMX', 7), 
             ('OAX', 14), ('CMX', 10), ('PUE', 12)]

estados_empleados_unicos = list(set(emp[0] for emp in empleados))
print(estados_empleados_unicos)

estados_empleados = {}

for emp in empleados:
    if emp[0] in estados_empleados:
        estados_empleados[emp[0]] += emp[1]
    else:
        estados_empleados[emp[0]] = emp[1]
print(estados_empleados)
````
