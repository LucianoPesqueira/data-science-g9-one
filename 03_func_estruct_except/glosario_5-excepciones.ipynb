````python
notas = {"Juan":[8.0, 9.0, 10.0], "Maria":[9.0, 7.0, 6.0], "Jose":[3.4, 7.0, 7.0], "Claudia":[5.5, 6.6, 8.0], "Ana":[6.0, 10.0, 9.5], "Jorge":[5.5, 7.5, 9.0], "Julia":[6.0, 8.0, 7.0], "Pedro":[3.0, 4.0, 6.0]}
````

````python
try:
    nombre = input("Ingrese el nombre del alumno: ")
    resultado = notas[nombre]
except Exception as e:
    print(type(e), e)
#<class 'KeyError'> 'maria'
````

````python
try:
    nombre = input("Ingrese el nombre del alumno: ")
    resultado = notas[nombre]
except KeyError :
    print("Alumno no encontrado!")
else:
    print(resultado)
````

````python
try:
    nombre = input("Ingrese el nombre del alumno: ")
    resultado = notas[nombre]
except KeyError :
    print("Alumno no encontrado!")
else:
    print(resultado)
finally:
    print("La consulta ha concluido!")
````

````python
def promedio(lista: list=[0]) ->float:
    '''Funcion para calcular el promedio de notas de una lista  

    lista: list, default [0]  
      Lista con las notas para calcular el promedio  
    return = calculo: float  
      Promedio Calculado  
    '''
    calculo = sum(lista) / len(lista)
    
    if len(lista) > 4:
        raise ValueError("No puede poseer mas de 4 notas")
    
    return calculo
````

````python
notas = [6, 7, 8, 9, 10]
resultado = promedio(notas)
print(resultado)
````

````python
try:
    notas = [6, 7, 8, 9, 10]
    resultado = promedio(notas)
except TypeError:
    print("No fue posible calcular el promedio del estudiante, solo se admiten valores numericos")
except ValueError as e:
    print(e)
else:
    print(f"Las notas del estudiante son: {resultado}")
finally:
    print("La consulta a terminado!")
````

````python
empleados = {'José': 2000, 'Ana': 2200, 'Juan': 2500, 'Maria': 3800}

try:
    aumento = list(map(lambda x: x[1] * 1.1, empleados.items()))
    aumento2 = list(map(lambda x: x * 1.1, empleados.values()))
except ValueError as e:
    print(type(e), f"error: {e}")
else:
    print(aumento)
    print(aumento2)
finally:
    print("Proceso concluido")
````
