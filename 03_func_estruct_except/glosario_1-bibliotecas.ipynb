```python
from random import sample

frutas = ["manzana", "banana", "uva", "pera", "mango", "coco", "sandia", "fresa", "naranja", "maracuya", "kiwi", "cereza"]

ensalada_frutas = sample(frutas, 3)
print(f"La ensalada de frutas sera con: {ensalada_frutas}")
```

```python
from math import sqrt

numeros = [2, 8, 15, 23, 91, 112, 256]

for n in numeros:
    if sqrt(n) % 2 == 0:
        print(f"La raiz cuadrada de {n} que es un numero entero es: {sqrt(n)}")
```

```python
import math 

metro_cuadrada = 25.00
radio = float(input("Ingrese el radio del area circular: "))
area = math.pi * (radio ** 2)

print(f"El monto a pagar ${area * metro_cuadrada}")
```
