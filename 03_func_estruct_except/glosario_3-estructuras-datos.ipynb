# Estructuras de datos

### Listas de listas
- [[a1, a2,...an], [b1, b2,...bn],...,[n1, n2,...,nn]]

```python
notas_grupo = ["juan", 8.0, 9.0, 10.0, "Maria", 9.0, 7.0, 6.0, "Jose", 3.4, 7.0, 7.0, "Claudia", 5.5, 6.6, 8.0, "Ana", 6.0, 10.0, 9.5]
notas = []
nombres = []

for i in range(len(notas_grupo)):
    if i % 4 == 0:
        nombres.append(notas_grupo[i])
    else:
        notas.append(notas_grupo[i])

notas_separadas = []
for i in range(0, len(notas), 3):
    notas_separadas.append([notas[i], notas[i+1], notas[i+2]])
    
print(notas_separadas)
```

### Listas de tuplas
- [(a1, a2,...an), (b1, b2,...bn),...,(n1, n2,...,nn)]

```python
from random import randint

nombres = ["Juan", "Maria", "Jose", "Claudia", "Ana"]

def genera_numero():
    return randint(0, 999)

codigo_estudiante = []

for i in range(len(nombres)):
    codigo_estudiante.append((nombres[i], nombres[i][0]+str(genera_numero())))
    
print(codigo_estudiante)
```
