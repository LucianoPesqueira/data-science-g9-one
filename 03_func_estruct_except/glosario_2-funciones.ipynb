````python
#funciones
def promedio(nota_1, nota_2, nota_3):
    calculo = (nota_1 + nota_2 + nota_3) / 3
    print(calculo)
    
promedio(2, 6, 1)
promedio(nota_1 = 10, nota_2 = 4, nota_3 = 6)
````

````python
# RETURN

notas = [8.5, 9.0, 7.5, 4, 9.5, 6.5]

def promedio(lista):
    resultado = sum(lista) / len(lista)
    return resultado

print(promedio(notas))#7.5
````

````python
notas = [8.5, 9.0, 7.5, 4, 9.5, 6.5]

def promedio(lista):
    resultado = sum(lista) / len(lista)
    if resultado >= 7:
        situacion = "Aprobado"
    else:
        situacion = "Reprobado"
    return (resultado, situacion)

promedio(notas)#(7.5, 'Aprobado')

resultado, situacion = promedio(notas)

print(f"El estudiante obtuvo un puntaje de: {resultado} y su situacion es: {situacion}")
````

## Funciones Lambda

- Son funciones anonimas (no necesita definirla)
- Pueden ser escritas en una linea
- ```lambda <variable>: <expresion>```
- ```map(<lambda function>, <iterador>)```

* Type Hint
- sintaxis utilizada para indicar el tipo de dato esperado de un parametro o el retorno de una funcion.  
```def <nombre>(<param>: <tipo_param>) -> <tipo_retorno>:```  
  ```<instrucciones>```  
  ```return resultado```

````python
nota_1 = float(input("Ingrese una nota"))

def cualitativa(nota):
    return nota + 0.5

cualitativa(nota_1)
````

````python
nota = float(input("Ingrese una nota"))

cualitativa = lambda x: x + 0.5

cualitativa(nota)
````

````python
#ejercicio1
nota1 = float(input("Ingrese la primera nota: "))
nota2 = float(input("Ingrese la segunda nota: "))
nota3 = float(input("Ingrese la tercera nota: "))

promedio = lambda x, y, z: (x*3+y*2+z*5)/10

promedio(nota1, nota2, nota3)
````

````python
#map(lambda)
#notas
notas = (6.0, 7.0, 9.0, 5.5, 8.0) #-> para aplicar a iterables tengo que usar una funcion map, filter, reduce
cualitativo = 0.5

notas_actualizadas = list(map(lambda x: x + cualitativo, notas))

print(notas_actualizadas)
````

````python
#Ejercicio Convirtiendo temperaturas
temp_celsius = [0, 25, 37, 78, 100]

#temp_fahrenheit = temp_celsius * 9/5 + 32
temp_fahrenheit = list(map(lambda x: (x * 9/5) + 32, temp_celsius))

print(temp_fahrenheit)
````

````python
#5
# mayor nota
# menor nota
# media
# situación (Aprobado(a) o Reprobado(a))
# Uso de la función
# Mostrar: El estudiante obtuvo una media de `media`, con la mayor nota de `mayor` puntos y la menor nota de `menor` puntos y fue `situacion`.)

def rendimiento(notas):
    mayor = max(notas)
    menor = min(notas)
    media = round(sum(notas) / len(notas), 2)
    situacion = "Aprobado" if media >= 6 else "Reprobado"
    return f"El estudiante obtuvo una media de {media}, con la mayor nota de {mayor} puntos y la menor nota de {menor} puntos y fue {situacion}."

lista = [3, 4.5, 7, 9, 8.5, 6.5]
rendimiento(lista)
````
