## Desafio

1. Verifica si la base de datos contiene datos nulos y, en caso de tenerlos, realiza el tratamiento de estos datos nulos de la manera que consideres más coherente con   la situación.

2. Los estudiantes "Alicia" y "Carlos" ya no forman parte del grupo. Por lo tanto, elimínalos de la base de datos.

3. Aplica un filtro que seleccione solo a los estudiantes que fueron aprobados.

4. Guarda el DataFrame que contiene solo a los estudiantes aprobados en un archivo CSV llamado "alumnos_aprobados.csv".

```python
import pandas as pd
datos = pd.read_csv('alumnos.csv', sep=",")
```

```python
#1
datos.isnull()#-> me muestra la tabla con los null como True
```

```python
dt = datos.fillna(0) #-> ej 1 : cambio nan por 0
dt
```

```python
#2 eliminar a 'Alicia' y 'Carlos'
dt_remove = dt.query('Nombre == "Alicia" | Nombre == "Carlos"').index
dt.drop(dt_remove, axis=0, inplace=True)
```

```python
#3 primero arreglo los datos, ya que Nadia es Verdadero en vez de True
dt["Aprobado"] = dt['Aprobado'].replace({'Verdadero':True, 'True':True, 'False':False})
dt
```

```python
#3
filtro1 = dt["Aprobado"] == True
dt = dt[filtro1]
```

```python
#4 guardar como alumnos_aprobados.csv
dt.to_csv('alumnos_aprobados.csv', index=False, sep=';')
```

```python
#extra - los estudiantes con nota 7.0 tienen punto extra y llegan a 8.0
new_dt = pd.read_csv('alumnos_aprobados.csv', sep=';')
```

```python
new_dt.dtypes #-> me muestra los tipos de datos para saber si la nota es entera o flotante
```

```python
#obtengo los estudiantes con nota 7
filtro_nota = new_dt["Nota"] == 7.0
new_dt["Nota"] = new_dt["Nota"].replace({7.0: 8.0})
new_dt
```

```python
# ahora lo guardo con la nota cambiada:
new_dt.to_csv('alumnos_aprobados.csv', index=False, sep=';')
```
